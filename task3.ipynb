```python
import math

lam = 5

expected_value = lam
variance = lam
std_dev = math.sqrt(lam)

print("Poisson Distribution with λ = 5")
print("-------------------------------")
print(f"Expected value E[X] = {expected_value}")
print(f"Variance Var(X) = {variance}")
print(f"Standard deviation = {round(std_dev, 4)}")
```
