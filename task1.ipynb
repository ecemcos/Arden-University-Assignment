```python
import numpy as np

A = np.array([[1.2, 0.1],
              [0.05, 1.1]])

eigenvalues, eigenvectors = np.linalg.eig(A)

order = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[order]
eigenvectors = eigenvectors[:, order]

v1 = eigenvectors[:, 0] / eigenvectors[1, 0]
v2 = eigenvectors[:, 1] / eigenvectors[1, 1]

l1 = round(eigenvalues[0], 3)
l2 = round(eigenvalues[1], 3)

v1_x = round(v1[0], 3)
v2_x = round(v2[0], 3)

print("Eigenvalues")
print(f"Eigenvalue for λ1 ≈ {l1}")
print(f"Eigenvalue for λ2 ≈ {l2}")

print("\nEigenvectors")
print(f"Eigenvector for v1 ∝ ({v1_x}, 1)")
print(f"Eigenvector for v2 ∝ ({v2_x}, 1)")


```
